```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

launches = pd.read_csv('launches.csv')
agencies = pd.read_csv('agencies.csv')
```

```python
launches
```

```python
agencies
```

# A. Launches by Agency Type 

```python
launches['agency_type'].value_counts()
```

```python
year_agency_type = launches.groupby(['launch_year','agency_type'])['launch_year'].count().unstack().fillna(0)
```

```python
year_agency_type
```

```python
#Create the line plot
sns.relplot(data=year_agency_type, kind='line',height=8, aspect= 1.3)


plt.xlabel('Launch Year',fontsize=15)
plt.ylabel('Number of Launches',fontsize=15)
plt.title('Space Launches by Agency Type',fontsize=17)

plt.show()
```

# Space Launches by Country

```python
launches['state_code'].value_counts()
```

```python
launches['state_code'] = launches['state_code'].str.replace('SU','RU')
```

```python
launches['category'] = launches['category'].replace({'O': 1,'F': 0})
```

```python
top_5_countries = launches[launches['state_code'].isin(['RU','US','CN','F','J'])]
```

```python
year_countries = top_5_countries.pivot_table(index='launch_year',columns='state_code',fill_value = 0,aggfunc='size')
```

```python
year_countries
```

```python
sns.relplot(data=year_countries, kind='line',height=7, aspect=1.5)
plt.xlabel('Lanuch Year', fontsize=15)
plt.ylabel('Number of Launches', fontsize=15)
plt.title('Space Launches By Country',fontsize=17)
```

```python
plt.figure(figsize=(11,9))
names = ['RU','US','CN','F','J']
for i in range(5):
    year_countries[names[i]].plot(kind='area',label= names[i])
plt.xlabel('Lanuch Year', fontsize=15)
plt.ylabel('Number of Launches', fontsize=15)
plt.title('Space Launches By Country',fontsize=17)
plt.axvline(x= 1969, label= 'Apollo 11')
plt.axvline(x= 1957, label= 'Sputnik 1', color = 'black')
plt.legend()
plt.show()
```

# Which country has more Failed Launches? 

```python
success_failure = top_5_countries.pivot_table(index='state_code',columns='category', aggfunc='size')
success_failure.columns = ['failure','success']
success_failure
```

```python
success_failure['failure_percentage'] = round((success_failure['failure'] /success_failure.sum(axis=1))*100, 2)

success_failure.reset_index(inplace=True)

success_failure
```

```python
sns.catplot(x='state_code',y='failure_percentage', kind = 'bar',data=success_failure, height= 7, aspect=1)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Failure Percentage', fontsize=14)
plt.title('Failure Percentage BY Country', fontsize=17)
plt.show()
```

# Private and Startups Agencies- 


```python
private_startups = launches[(launches['agency_type'] == 'private') | (launches['agency_type']== 'startup')]

private_startups['agency'].value_counts()
```

```python
private_startups = pd.merge(left= private_startups, right=agencies, 
on='agency', how='left',suffixes=('_ps','_agency'))

private_startups['name'].value_counts()
```

```python
plt.figure(figsize=(9,8))
private_startups['name'].value_counts()[:10].sort_values().plot(kind='barh')

plt.xlabel('Number Of Launches', fontsize=13)
plt.ylabel('Private And Startup Agencies', fontsize=13)
plt.title('Most Active Private/Startup Agencies', fontsize=15)
plt.show()
```

```python
top_6_names = private_startups['name'].value_counts()[:6].index
private_startups_6 = private_startups[private_startups['name'].isin(top_6_names)]
private_startups_6['name'].value_counts()
```

```python
year_agency_count = private_startups_6.pivot_table(index=['launch_year','name'],aggfunc='size').reset_index(name='count')

year_agency_count
```

```python
sns.relplot(x='launch_year', y='count',col='name',
            col_wrap =3,kind='line',data=year_agency_count)

plt.show()
```

# Type Of Space Vehicles used most by Countries - 

```python
russia = launches[launches['state_code'] == 'RU']

russia['type'].value_counts()
```

```python
only_top_15_names = russia['type'].value_counts()[:15].index

russia_15 = russia[russia['type'].isin(only_top_15_names)]

russian_vehicles = russia_15.pivot_table(index=['launch_year','type','agency_type'],aggfunc='size').reset_index(name='count')

russian_vehicles.head()
```

```python
plt.figure(figsize=(12,10))

sns.stripplot(y='type',x='launch_year', hue='agency_type',
              size= 8 ,data= russian_vehicles)

plt.xlabel('Launch Year', fontsize=13)
plt.ylabel('Vehicle Types', fontsize=13)
plt.title('Timeline Of Top 15 Russian Vehicles', fontsize=15)
plt.show()
```

# US Vehicles -

```python
US = launches[launches['state_code'] == 'US']

only_top_15_names = US['type'].value_counts()[:15].index

US_15 = US[US['type'].isin(only_top_15_names)]

US_15['type'].value_counts()
```

```python
US = launches[launches['state_code'] == 'US']
only_top_15_names = US['type'].value_counts()[:15].index
US_15 = US[US['type'].isin(only_top_15_names)]
US_15['type'].value_counts()
```

```python
us_vehicles = US_15.pivot_table(index=['launch_year','type','agency_type'],aggfunc='size').reset_index(name='count')
us_vehicles.head()
```

```python
plt.figure(figsize=(12,10))

sns.stripplot(y='type',x='launch_year', hue='agency_type',
              size= 8 ,data= us_vehicles)

plt.xlabel('Launch Year', fontsize=13)
plt.ylabel('Vehicle Types', fontsize=13)
plt.title('Timeline Of Top 15 US Vehicles', fontsize=15)
plt.show()
```
